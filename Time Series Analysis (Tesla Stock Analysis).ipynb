# Moving Average

```python
df_tesla[['Open','Open:30 days rolling']].plot(figsize=(12,5))
```

```python
##Simple Moving Average
##Evaluating 20,50,200 days moving averaage to get buying and selling signals
```

```python
df_tesla['Open'].plot(figsize=(15,6))
```

```python
df_tesla['Open:10 days rolling'] = df_tesla['Open'].rolling(window =10, min_periods =1 ).mean()
```

```python
df_tesla[['Open','Open:10 days rolling']].plot(xlim=['2021-01-01', '2022-01-01'],figsize=(15,6))
```

```python
df_tesla['Open:50 days rolling'] = df_tesla['Open'].rolling(window =50, min_periods =1 ).mean()
```

```python
df_tesla[['Open','Open:50 days rolling']].plot(xlim=['2021-01-01', '2022-01-01'],figsize=(15,6))
```

```python
df_tesla['Open:200 days rolling'] = df_tesla['Open'].rolling(window =200, min_periods =1 ).mean()
```

```python
df_tesla[['Open','Open:200 days rolling']].plot(xlim=['2021-01-01', '2022-01-01'],figsize=(15,6))
```

```python
df_tesla[['Open','Open:200 days rolling','Open:10 days rolling','Open:50 days rolling']].plot(xlim=['2020-01-01', '2022-01-01'],figsize=(15,6))
```

```python
##Cumulative Moving Average
df_tesla['Open'].expanding().mean().plot(figsize=(10,5))
```

```python
#EWMA = Exponencial Weighted Moving Average
##FORMULLA = ((CLOSE EMA - PREV EMA) * MULTIPLIER) + PREV EMA
```

```python
df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1,adjust= False).mean()
```

```python
df_tesla['EMA_0.3'] = df_tesla['Open'].ewm(alpha=0.3,adjust= False).mean()
```

```python
df_tesla[['Open','EMA_0.1','EMA_0.3']].plot(xlim=['2020-01-01', '2022-12-01'],figsize=(15,6))
```

```python
df_tesla['EMA_5days'] = df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open','EMA_5days','EMA_0.3','EMA_0.1']].plot(figsize = (12,5))
```
